## Using the matplotlib package

The matplotlib package is a powerful, very general, free plotting program that is driven by Python. To get an idea of its abilities check out the plot gallery at

https://matplotlib.org/gallery.html

with thumbnails of plots, and matplotlib/Python code.

Besides the vast number of plot types, the ability to combine computation and plotting within the same program is very powerful. It also helps you to automate your entire workflow of data analysis and presentation.


```python
# a jupyter notebook specific command to allow plots inside the notebook
#
# import the package we need
import matplotlib.pyplot as plt
```

### Line plot

```python
# 
# this code is just to fake some data to plot
import numpy as np
x = np.linspace(-5,5,50) # equally spaced values for the x-axis
y = np.exp(-1.*x**2) # Gaussian
```

The matplot commands- only the minimal subset illustrated


```python
plt.figure() 
plt.plot(x,y) # line plot in green
#
# labelling
#plt.xlabel('x')
#plt.ylabel('y')
#plt.title('example line plot')
plt.show()
```

### box and whisker plot

```python
#
# fake up two data sets
data1 = np.linspace(1.5,2.7,13)
data2 = np.random.rand(10)
#
# put them all into one list
dataAll = [data1,data2]
# print(dataAll)
plt.figure()
plt.boxplot(dataAll)
#
# labelling
plt.xlabel('some data')
plt.ylabel('results')
plt.xticks([1, 2], ['data1', 'data2'])
plt.title('box and whisker plot')
plt.show()
```

### scatter plot

```python
#
# fake some data
np.random.seed(7777777)
x = np.random.rand(20)
y = np.random.rand(20)
plt.figure() 
plt.scatter(x,y,color='r',marker='s') # scatter plot in red squares
plt.scatter(y,x,color='b',marker='o') # green cirlces
#
# labelling
plt.xlabel('y')
plt.ylabel('x')
plt.xlim(0.,1.)
plt.ylim(-0.5,1.5)
plt.title('example scatter plot')
plt.show()
```
